```python
### importing libraries and Functions
import joblib
from pandas import read_csv,DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from category_encoders import TargetEncoder

data3 = read_csv('/content/Fossil Age test_data.csv') #Data Reading
 # Encoding the Data through Map Function
data3['paleomagnetic_data']= data3['paleomagnetic_data'].map({'Normal polarity':0,'Reversed polarity':1 })
data3['stratigraphic_position']= data3['stratigraphic_position'].map({'Bottom':0,'Middle':1,'Top':2 })
data3['geological_period']= data3['geological_period'].map({'Cretaceous':8,'Cambrian':0,'Permian':5, 'Devonian':3, 'Jurassic':7, 'Neogene':10, 'Triassic':6, 'Paleogene':9, 'Ordovician':1, 'Carboniferous':4, 'Silurian':2  })
data3['inclusion_of_other_fossils'] = data3['inclusion_of_other_fossils'].map({True:1, False:0})
 # Features Selection (Dividing X and Y)
X1 = data3.drop(['age'], axis = 1)
Y1 = data3['age']
print(X1.shape)
print(Y1.shape)
 #Target Encoding for remaining Catagrocial Features
Target_Encoder2 = TargetEncoder(cols=['surrounding_rock_type' ]).fit(X1,Y1)
data4= Target_Encoder2.transform(X1)
print(data4.info())
data4.head(10)
X1_= StandardScaler().fit_transform(data4)        # Scaling Training Data
DataFrame(X1_).head()
# Importing the pickel which is based on onle of our models
new_model = joblib.load("l1fossilmodel.pkl")
# Predicting the Output of New Data using the model
x = X1_
predict = new_model.predict(x)[:]   # using the pkl file for prediction
data5 = read_csv('/content/Fossil Age test_data.csv')   # Importing New Data
data5['predicted age'] = predict # adding new column based on predictions
data5.head(10)
```
